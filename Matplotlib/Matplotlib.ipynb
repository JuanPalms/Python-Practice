#### Basic x,y plot

```python
x_data=np.arange(0,101,10)
y_data=x_data**2
noise=np.random.randint(-10,10,11)
#element wise addition
y_data=np.add(y_data,noise)
```

```python
fig=plt.figure(1, figsize=(5,5))
#customize marker, line
plt.plot(x_data,y_data,
         marker='o',markersize=10,mfc='w',
         linestyle='none')  #linestyle='none'>just see your x,y data ploted
```

```python
fig=plt.figure(1, figsize=(5,5))
#customize marker, line
plt.plot(x_data,y_data,
         marker='o',markersize=10,mfc='w',
         linestyle='-.',label='data_set 1') #we pass the argument label in order to add it but does not show nedd legend
plt.legend(loc='lower right') #loc allows to locate the label whenever you want
plt.show()
```

```python
fig=plt.figure(1, figsize=(5,5))
#customize marker, line
plt.plot(x_data,y_data,
         marker='o',markersize=10,mfc='w',
         linestyle='-.',label='data_set 1') 
plt.plot(x_data,y_data+500,
         marker='o',markersize=10,mfc='w',
         linestyle='--',label='data_set 2')
plt.legend(loc='upper left') #loc allows to locate the label whenever you want
plt.show()
```

```python
fig=plt.figure(1, figsize=(5,5))
#customize marker, line
plt.semilogy(x_data,y_data,
         marker='o',markersize=10,mfc='w',
         linestyle='-.',label='data_set 1') #we can change the scale of the y label to logaritmic with semilogy
plt.semilogy(x_data,y_data+500,
         marker='o',markersize=10,mfc='w',
         linestyle='--',label='$y=x^2+500$') #we can set python to math mode in the labels with $..$
#putting labels in axis
plt.xlabel('x variable')
plt.ylabel('y variable')
plt.legend(loc='upper left') 
plt.show()
```

#### Categorical plots

```python
rings={'elves':3, 'men':9, 'dwarves':7, 'sauron':1}
```

```python
lists_rings=sorted(rings.items(), key=lambda item: item[1]) #if i want it to rank alphabetically, use item[0]
x,y=zip(*lists_rings)
```

```python
plt.bar(x,y) #verticaly
```

```python
plt.barh(x,y) #horizontaly
```

#### Pie chart with categorical data

```python
plt.pie(y,labels=x)
```

```python
#in order to explode (destacar una o mas categorias)
explode=(0.5,0.2,0,0.05)
plt.pie(y,labels=x , explode=explode)
```

### Scatter plots

```python
x_data=np.arange(0,50,1)
y_data=np.arange(0,50,1)
colors=np.arange(0,50,1)
area=np.random.rand(50)**2*1000
fig=plt.figure(1,figsize=(5,5))
plt.scatter(x_data,y_data,c=colors,s=area, alpha=0.8)
```

### Error plots

```python
x_data= np.arange(0,2*np.pi+0.1,np.pi/6)
y_data= np.sin(x_data)
y_error=np.linspace(0,2,13)
```

```python
fig=plt.figure(1,figsize=(5,5))
plt.errorbar(x_data, y_data, yerr=y_error,
            marker='o',mfc='w',capsize=5, markersize=10)
```
